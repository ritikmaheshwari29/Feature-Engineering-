**Process of determining the attitude or emotion of the writer is called "Sentiment Analysis"**

## Finding the Length, Polarity, Subjectivity

```python
data['length'] = data['verified_reviews'].apply(len)
```

```python
data.head()
```

### Finding a sentiment of the review:
- **Textblob** has a function called **sentiment** which returns two properties **"polarity" and "subjectivity"**.

## Text Polarity

- It is the expression that determines the sentimental aspect of an opinion.
- In textual data, the result of sentiment analysis can be determined for each entity in the sentence, document or sentence.
- The sentiment polarity can be determined as positive, negative and neutral.

- Polarity lie between -1 to +1, where +1 means positive sentiment and vice cerca.
    

```python
def get_polarity(text):
    textblob = TextBlob(str(text.encode('utf-8')))
    pol = textblob.sentiment.polarity
    return pol
    
data['Polarity'] = data['verified_reviews'].apply(get_polarity)
```

```python
data.head()
```

## Text Subjectivity
- In natural language, subjectivity refers to expression of opinions, evaluations , feelings, and speculations and thus incorporates sentiment. 
- Subjective text is further classified with sentiment or polarity.

- Subjectivity lies between **0(a very objective statement) to 1(a very subjective statement)**.

```python
def get_subjectivity(text):
    textblob = TextBlob(str(text.encode('utf-8')))
    subj = textblob.sentiment.subjectivity
    return subj
    
data['Subjectivity'] = data['verified_reviews'].apply(get_subjectivity)

data.head()
```
