## Finding the words,characters, and punctuation count

```python
import pandas as pd
```

```python
data = pd.read_csv('amazon_alexa.tsv',delimiter='\t')

data.shape
```

```python
data.head()
```

```python
# calculate the character count in the reviews
data['char_count'] = data['verified_reviews'].apply(len)
```

```python
data.head()
```

```python
# calculating the word count
data['word_count'] = data['verified_reviews'].apply(lambda x: len(x.split()))
data.head()
```

```python
# calculating word density
data['word_density'] = data['char_count'] / (data['word_count']+1)
data.head()
```

```python
# importing list of punctuation
import string
punctuation = string.punctuation

# calculating punctuation count
data['punctuation_count'] = data['verified_reviews'].apply(lambda x: len("".join(_ for _ in x if _ in x if _ in punctuation)))
```

```python
## lets summarize the Newly Created Features
data[['char_count','word_count','word_density','punctuation_count']].describe()
```
