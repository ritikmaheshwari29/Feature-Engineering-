## Decomposing Categorical Features

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('fifa.csv')

data.shape
```

```python
pd.set_option('max_columns',89)
data.head()
```

```python
data['Nationality'].nunique()
```

```python
pd.set_option('max_rows',164)
data['Nationality'].value_counts()
```

```python
def decompose_categorical(x):
    if x=='England':
        return 'England'
    elif x=='Germany':
        return 'Germany'
    elif x=='Spain':
        return 'Spain'
    elif x=='Argentina':
        return 'Argentina'
    elif x=='France':
        return 'France'
    else:
        return 'Others'
    
data['Nationality'] = data['Nationality'].apply(decompose_categorical)
data['Nationality'].value_counts()
```
