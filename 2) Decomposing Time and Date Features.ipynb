## Decomposing Time and Date Features

- Date variables are a special type of categorical variable.

- From date variable we can extract basic features like:

    - Year
    - Month
    - Day
    - Day of the week
    - Time difference etc.

```python
import pandas as pd
import numpy as np
import datetime
```

```python
# periods=15 means craeting 15 rows
time_series = pd.date_range('1/1/2020', periods=15)

df = pd.DataFrame({'date':time_series})
df
```

```python
df.dtypes
```

```python
# Converting to DateTime type
df['date'] = pd.to_datetime(df['date'])
```

Note: **SERIES.DT is used to access the values of the series of a Datetime variable.**

```python
# Extracting year, month, day

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```
